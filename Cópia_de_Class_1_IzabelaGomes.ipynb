**1.- Conversor de dias para segundos**

Pea ao usurio para inserir um nmero de dias e converta-o em segundos.

````python
# Inserir o nmero de dias
dias = int(input("Digite o nmero de dias: "))

# Calcula o nmero de segundos em um dia (1 dia = 24 horas, 1 hora = 60 minutos, 1 minuto = 60 segundos)
segundos_por_dia = 24 * 60 * 60

# Calcula o total de segundos multiplicando o nmero de dias pelos segundos em um dia
total_segundos = dias * segundos_por_dia

#Resultado
print(f"{dias} dias equivalem a {total_segundos} segundos.")
````

**2.- Desconto em compras**

Crie um programa que solicite ao usurio o valor total de sua compra. Aplique um desconto de 10% no total da compra. Imprima o valor aps o desconto.

````python

valor_total = float(input("Digite o valor total da sua compra: R$ "))

# O desconto como 10% do valor total. Multiplicamos o valor total por 0.10 (10%).
desconto = valor_total * 0.10

# Calcula o valor final aps aplicar o desconto
valor_final = valor_total - desconto

# Resultado
print(f"O valor aps o desconto de 10% :  R$ {valor_final} ")
````

**5.- Calculadora de ndice de Massa Corporal (IMC)**


Escreva um programa que solicite ao usurio o peso (em quilogramas) e a altura (em metros) e calcule o IMC.

IMC = peso / (altura ^ 2)

````python
#Inserir o peso em quilogramas
peso = float(input("Digite o seu peso em quilogramas (kg): "))

# Para inserir a altura em metros
altura = float(input("Digite a sua altura em metros (m): "))

# Calcula o IMC usando a frmula IMC = peso / (altura ^ 2)
imc = peso / (altura ** 2)

# Exibe o resultado para o usurio
print(f"Seu ndice de Massa Corporal (IMC) : {imc:.2f}")
````

**6.- Hipotenusa**

Desenvolva um programa que solicite ao usurio que insira os comprimentos dos dois catetos de um tringulo retngulo e calcule o comprimento da hipotenusa.

Dica: Use a biblioteca matemtica.

````python
#importado da biblioteca de math (necessria para encontrar a hipotenusa, usamos math.sqrt() )
import math


#Ao inserir os valores dos catetos a e b. A funo float converte esses valores para nmeros de ponto flutuante (nmeros reais).

# Solicitar para inserir o comprimento do primeiro cateto
cateto1 = float(input("Digite o comprimento do primeiro cateto: "))

# Solicitar para inserir o comprimento do segundo cateto
cateto2 = float(input("Digite o comprimento do segundo cateto: "))

# Calcula o quadrado da hipotenusa usando o Teorema de Pitgoras
hipotenusa_quadrado = cateto1**2 + cateto2**2

# Calcula a hipotenusa tirando a raiz quadrada da soma dos quadrados dos catetos
hipotenusa = math.sqrt(hipotenusa_quadrado)

# Exibe o resultado para o usurio
print(f"O comprimento da hipotenusa : {hipotenusa}")
````

**8.- Criao de e-mail**

Solicite ao usurio que insira seu nome e sobrenome. Em seguida, gere um e-mail concatenando a inicial do nome com o sobrenome seguido de @gmail.com. O e-mail deve estar todo em letras minsculas.

Exemplo:

nome: Gustavo

Sobrenome: Lavadenz

e-mail: glavadenz@gmail.com

````python
# Solicita ao usurio que insira seu nome
nome = input("Digite seu nome: ")

# Solicita ao usurio que insira seu sobrenome
sobrenome = input("Digite seu sobrenome: ")

# Gera o e-mail concatenando a inicial do nome com o sobrenome e adicionando "@gmail.com"
email = f"{nome[0].lower()}{sobrenome.lower()}@gmail.com"

# Exibe o e-mail gerado
print(f"O seu e-mail : {email}")
````
